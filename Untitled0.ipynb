```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
import pandas as pd
import datetime
%matplotlib inline

df = read_csv('dengue-data.csv', engine = 'python')

timeseries = df["Date"]

t = []
nt = []

dx = []

for i in range(len(timeseries)):
    time  = timeseries[i]
    k = time.split('-')
    s = k[0]+"-"+k[1]+'-01'
    t.append(s)
    if s not in nt:
        nt.append(s)
        dx.append(k[0]+'-'+k[1])
df['Date'] = t
k = df.groupby(['Date']).mean()
print k
```

```python
import datetime as dt
x = [dt.datetime.strptime(d, '%Y-%m-%d').date() for d in nt]
y = []

# x = x[:36]
for i in range(len(k)):
    y.append(k["India"][i])
plt.subplots(figsize=(20,10))
# y = y[:36]

plt.plot(x,y)
plt.show()
```
