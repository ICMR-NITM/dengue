```python
from keras.layers.core import Dense, Activation, Dropout

from keras.layers.recurrent import LSTM
from keras.models import Sequential

import lstm, time
```

```python


X_train, y_train, X_test, y_test = lstm.load_data('newprocessed.csv', 50, True)

print len(X_train)

print len(X_test)
```

```python

#Step 2 Build Model
model = Sequential()

model.add(LSTM(
    input_dim=1,
    output_dim=50,
    return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(
    100,
    return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(
    output_dim=1))
model.add(Activation('linear'))

start = time.time()
model.compile(loss='mse', optimizer='rmsprop')
print 'compilation time : ', time.time() - start
```

```python
model.fit(
    X_train,
    y_train,
    batch_size=512,
    nb_epoch=100,
    validation_split=0.05)
```

```python
print len(X_test)
predictions = lstm.predict_sequences_multiple(model, X_test, 10, 10)

print predictions

lstm.plot_results_multiple(predictions, y_test, 3)
```
